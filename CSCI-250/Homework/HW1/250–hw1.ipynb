```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

# loading data
df = pd.read_csv('../../Data/avocado-updated-2020.csv')
```

##### a) Outputs a bar chart with the average year wise price of regular and organic avocados for each of the six years (two bars for each year)

```python
# calculating average price by year and type
yearly_price = df.groupby(['year', 'type'])['average_price'].mean()
# tidying data based on type
yearly_price = yearly_price.unstack()

# plotting
yearly_price.plot(kind='bar', figsize=(10, 5))
plt.title("Average Price by Year")
plt.ylabel("Average Price")
plt.xlabel("Year")
plt.show()
```

##### b) Outputs a line chart with two lines: the total volume of regular avocados sold per year for each of the six years, and the total volume of organic avocados sold per year for each of the six years.

```python
# calculating average price by region and type
yearly_volume = df.groupby(['year', 'type'])['total_volume'].sum()
yearly_volume = yearly_volume.unstack()

# plotting
plt.figure(figsize=(10, 5))

plt.plot(yearly_volume.index, yearly_volume['conventional'], label='Conventional', marker='o')
plt.plot(yearly_volume.index, yearly_volume['organic'], label='Organic', marker='x')

plt.title('Total Volume Sold by Year')
plt.xlabel('Year')
plt.ylabel('Total Volume')
plt.legend()
plt.show()
```
