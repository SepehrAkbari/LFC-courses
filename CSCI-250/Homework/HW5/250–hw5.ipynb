```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../Data/main/Fish.csv')
df.head()
```

## Question 1

Setting `Width` as independent variable and `Length` as dependent variable, and throwing the rest away.

```python
df = df[['Width', 'Weight']]

X = df[['Width']]  # Independent
y = df['Weight']   # Dependent
```

## Question 2 & 3

Fitting a linear, quadratic, and cubic model to the data, and visualizing the results.

**Explenation:**

The idea is to see all models as polynomials. A linear equation is a polynomial of degree 1, a quadratic equation is a polynomial of degree 2, and a cubic equation is a polynomial of degree 3. The general form of a polynomial of degree n is: 

$$y = a_0 + a_1x + a_2x^2 + a_3x^3 + ... + a_nx^n$$

So, we'll use the `PolynomialFeatures` class to generate the polynomial features for the models, before applying the `LinearRegression` class to fit the models to the data.

```python
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
```

1. **Linear Model**

```python
poly1 = PolynomialFeatures(degree=1)
X_poly1 = poly1.fit_transform(X)
lin_model = LinearRegression()
lin_model.fit(X_poly1, y)

y_lin_pred = lin_model.predict(poly1.transform(x_range))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='tab:blue', label='Data', alpha=0.75)
plt.plot(x_range, y_lin_pred, color='tab:red', linewidth=2, label='Linear Fit')
plt.xlabel('Width')
plt.ylabel('Weight')
plt.title('Linear Fit')
plt.legend()
plt.show()
```

2. **Quadratic Model**

```python
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)
quad_model = LinearRegression()
quad_model.fit(X_poly2, y)

y_quad_pred = quad_model.predict(poly2.transform(x_range))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='tab:blue', label='Data', alpha=0.75)
plt.plot(x_range, y_quad_pred, color='tab:red', linewidth=2, label='Quadratic Fit')
plt.xlabel('Width')
plt.ylabel('Weight')
plt.title('Quadratic Fit')
plt.legend()
plt.show()
```

3. **Cubic Model**

```python
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X)
cubic_model = LinearRegression()
cubic_model.fit(X_poly3, y)

y_cubic_pred = cubic_model.predict(poly3.transform(x_range))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='tab:blue', label='Data', alpha=0.75)
plt.plot(x_range, y_cubic_pred, color='tab:red', linewidth=2, label='Cubic Fit')
plt.xlabel('Width')
plt.ylabel('Weight')
plt.title('Cubic Fit')
plt.legend()
plt.show()
```
