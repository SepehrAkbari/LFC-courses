## Exercise 1

Compute and plot the quadric approximation of the function $\frac{1}{3x-x^2}$ centered at $1$. Use it to approximate values of $x$ in $[1,3]$

````python
def quadrApprox(function, center, lower_bound, upper_bound):
    x = symbols('x')
    f = function
    a = center

    fa = f.subs(x, a)
    f_prime = diff(f, x)
    fa_prime = f_prime.subs(x, a)
    f_double_prime = diff(f_prime, x)
    fa_double_prime = f_double_prime.subs(x, a)

    def taylorApprx(n):
        return fa + fa_prime * (n - a) + (fa_double_prime / 2) * (n - a) ** 2

    x_values = np.linspace(lower_bound, upper_bound - 0.01, 100)  
    f_values = [f.subs(x, n) for n in x_values]
    apprx_values = [taylorApprx(n) for n in x_values]

    plt.plot(x_values, f_values, label='f(x)')
    plt.plot(x_values, apprx_values, label='Approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Quadratic Approximation of f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

x = symbols('x')
f = 1 / (3 * x - x**2)
c = 1
lb = 1
ub = 3 

quadrApprox(f, c, lb, ub)
````
