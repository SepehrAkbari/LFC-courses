```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, exposure
```

```python
person = io.imread('person.jpg')
place = io.imread('place.jpg')
print(person.shape)
print(place.shape)
io.imshow(person)
```

```python
h = exposure.histogram(person, channel_axis=2)
h
```

```python
h[0][2].shape
```

```python
hist = np.concatenate([h[0][0],h[0][1],h[0][2]])
plt.plot(hist)
```

```python
# r - 65:111
# g - 193:253
# b - 90:133

output = np.copy(place)
for row in range(person.shape[0]):
    for col in range(person.shape[1]):
        if(not(64<=person[row, col, 0]<=112) or 
           not(192<=person[row, col, 1]<=254) or
           not(89<=person[row, col, 2]<=134)):
            output[row,col,:] = person[row,col,:]
            
            
plt.imshow(output)
```
