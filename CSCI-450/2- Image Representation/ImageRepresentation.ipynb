```python
import numpy as np
from skimage import data, color, io
import matplotlib.pyplot as plt
```

```python
img = data.astronaut()
print(img.shape)
print(type(img))
print(type(img[0,0,0]))
```

```python
plt.imshow(img)
```

```python
fig, (whole, r, g, b) = plt.subplots(ncols=4, figsize=(16, 16))

whole.imshow(img)
whole.axis('off')

r.imshow(img[:,:,0], cmap='gray')
r.axis('off')

g.imshow(img[:,:,1], cmap='gray')
g.axis('off')

b.imshow(img[:,:,2], cmap='gray')
b.axis('off')
```

```python
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]

g1 = red/3 + green/3 + blue/3
g2 = 0.2126*red + 0.7152*green + 0.0722*blue
g3 = 0.299*red + 0.587*green + 0.114*blue
g4 = 0.2627*red + 0.6780*green + 0.0593*blue
#g5 = 0.75*red + 0.20*green + 0.05*blue

fig, (gray1, gray2, gray3, gray4) = plt.subplots(ncols=4, figsize=(16, 16))

gray1.imshow(g1,'gray')
gray1.axis('off')

gray2.imshow(g2,'gray')
gray2.axis('off')

gray3.imshow(g3,'gray')
gray3.axis('off')

gray4.imshow(g4,'gray')
gray4.axis('off')
```

```python
g = color.rgb2gray(img)
plt.imshow(g,'gray')
print(g.shape)
```

```python
i = io.imread('colors.jpg')
y = color.rgb2gray(i)
io.imsave('grayscale.jpg',np.uint8(y*255))
```

```python
hsv = color.rgb2hsv(img)

fig, (h, s, v) = plt.subplots(ncols=3, figsize=(16, 16))

h.imshow(hsv[:,:,0], cmap='gray')
h.axis('off')

s.imshow(hsv[:,:,1], cmap='gray')
s.axis('off')

v.imshow(hsv[:,:,2], cmap='gray')
v.axis('off')
```

```python
hsv = color.rgb2(img)

fig, (h, s, v) = plt.subplots(ncols=3, figsize=(16, 16))

h.imshow(hsv[:,:,0], cmap='gray')
h.axis('off')

s.imshow(hsv[:,:,1], cmap='gray')
s.axis('off')

v.imshow(hsv[:,:,2], cmap='gray')
v.axis('off')
```
