### Question 1

Load the “astronaut” image from the skimage data module and do the
following:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import data, filters, transform

image = data.astronaut()
```

1. Apply a Roberts edge detection filter on the three color components
separately and stack the three edge images back together to form a
new image. Display the original and new image side by side. [5]

```python
# applying roberts to each edge color (from skimage)
r = filters.roberts(image[:, :, 0])
g = filters.roberts(image[:, :, 1])
b = filters.roberts(image[:, :, 2])

# stacking the edges
out = np.stack([r, g, b], axis = -1)

# plotting
fig,(orig,filt)= plt.subplots(ncols=2, figsize=(16, 8))

orig.imshow(image)
orig.axis('off')

filt.imshow(out)
filt.axis('off')
```

2. Reduce the image length and width each to 25% of the original, and
scale it back up to the original size. Display the original and modified
images side by side. (Hint: lookup skimage.transform.rescale()) [5]

```python
# reducing img by multimplying w and h by 0.25
# the rgb channels are stored in the last axis of the image array so we set channel_axis = -1
scale1 = transform.rescale(image, scale = 0.25, channel_axis = -1)

# increasing img by multiplying w and h by 4
scale2 = transform.rescale(scale1, scale = 4.0, channel_axis = -1)

# plotting
fig, (orig, scl2) = plt.subplots(ncols=2, figsize=(16, 8))

orig.imshow(image)
orig.axis("off")

scl2.imshow(scale2)
scl2.axis("off")
```
